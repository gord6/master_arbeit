```python
import pandas as pd
```

```python
config_datasheet = r'./results/preprocessed.xlsx'
```

```python
raw = pd.read_excel(config_datasheet)
df = pd.DataFrame(raw)
```

```python
new = pd.DataFrame(columns=[])

codes = df["Unique Code"].unique()

counter = 0

unique_code_list = []
point_list = []
max_point_list = []

for code in codes:
    tmp = df.loc[df["Unique Code"]== code]["Correctness"]
    for t in tmp:
        if t == True:
            counter += 1
    print(code, counter)
    unique_code_list.append(code)
    point_list.append(counter)
    max_point_list.append(10)
    counter = 0

new['Unique Code'] = unique_code_list
new['Punkte-Pretest-SoSci-Survey'] = point_list
new['Max Punkte-Pretest-SoSci-Survey'] = max_point_list
new
```
