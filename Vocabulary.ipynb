```python
import pandas as pd
import seaborn as sns
import collections
```

```python
config_datasheet = 'data/raw/Vokabeltraining.xlsx'
pre_datasheet = "results/data_of_all_removed.xlsx"
unique_code_datasheet = "results/behavioral/compare/compare.xlsx"
```

```python
raw_if = pd.read_excel(config_datasheet, sheet_name ="If")
raw_for = pd.read_excel(config_datasheet, sheet_name= "For")
raw_while = pd.read_excel(config_datasheet, sheet_name="While")
raw_do_while = pd.read_excel(config_datasheet, sheet_name="Do-While")
raw_pre = pd.read_excel(pre_datasheet)
raw_unique_codes = pd.read_excel(unique_code_datasheet)


df_if = pd.DataFrame(raw_if)
df_for = pd.DataFrame(raw_for)
df_while = pd.DataFrame(raw_while)
df_do_while = pd.DataFrame(raw_do_while)
df_pre = pd.DataFrame(raw_pre)
df_unique_codes = pd.DataFrame(raw_unique_codes)
df_unique_codes
```

```python
#unique_codes_if = df_if["Unique Code"].unique()
#unique_codes_for = df_for["Unique Code"].unique()
#unique_codes_while = df_while["Unique Code"].unique()
#unique_codes_do_while = df_do_while["Unique Code"].unique()
unique_codes_study = df_unique_codes["Unique Code"].dropna().unique()


df_vocabulary_if = pd.DataFrame()
df_vocabulary_for = pd.DataFrame()
df_vocabulary_while = pd.DataFrame()
df_vocabulary_do_while = pd.DataFrame()


for code in unique_codes_study:
    tmp_if = df_if.loc[df_if["Unique Code"]== code]
    tmp_for = df_for.loc[df_for["Unique Code"]== code]
    tmp_while = df_while.loc[df_while["Unique Code"]== code]
    tmp_do_while = df_do_while.loc[df_do_while["Unique Code"]== code]

    #tmp_post = df_post.loc[df_post["Unique Code"]==code]
    #df_result = df_result.append(tmp_pre)
    #df_result = df_result.append(tmp_post)
    df_vocabulary_if = df_vocabulary_if.append(tmp_if)
    df_vocabulary_for = df_vocabulary_for.append(tmp_for)
    df_vocabulary_while = df_vocabulary_while.append(tmp_while)
    df_vocabulary_do_while = df_vocabulary_do_while.append(tmp_do_while)
    
display(df_vocabulary_if)
path = "results/behavioral/new_vocabulary.xlsx"
writer = pd.ExcelWriter(path, engine='openpyxl')
df_vocabulary_if.to_excel(writer, 'if', index=False)
df_vocabulary_for.to_excel(writer, 'for', index=False)
df_vocabulary_while.to_excel(writer, 'while', index=False)
df_vocabulary_do_while.to_excel(writer, 'do-while', index=False)
writer.save()
#list_unique_codes = []

#for code in unique_codes_if:
#    list_unique_codes.append(code)

#for code in unique_codes_for:
#    list_unique_codes.append(code)
    
#for code in unique_codes_while:
#    list_unique_codes.append(code)
    
#for code in unique_codes_do_while:
#    list_unique_codes.append(code)
    
#for code in list_unique_codes:
#    list_unique_codes.append(code)

#display(unique_codes_study)   

#counter=collections.Counter(list_unique_codes)
#counter.most_common()
```

```python
mean_if = float(df_vocabulary_if["Punkte"].mean())
stddev_if = float(df_vocabulary_if["Punkte"].std())

mean_for = float(df_vocabulary_for["Punkte"].mean())
stddev_for = float(df_vocabulary_for["Punkte"].std())

mean_while = float(df_vocabulary_while["Punkte"].mean())
stddev_while = float(df_vocabulary_while["Punkte"].std())

mean_do_while = float(df_vocabulary_do_while["Punkte"].mean())
stddev_do_while = float(df_vocabulary_do_while["Punkte"].std())


print("if", mean_if, stddev_if)
print("for", mean_for, stddev_for)
print("while", mean_while, stddev_while)
print("do-while", mean_do_while, stddev_do_while)
```

```python
fehler_if = df_vocabulary_if["Fehler"].dropna()
fehler_for = df_vocabulary_for["Fehler"].dropna()
fehler_while = df_vocabulary_while["Fehler"].dropna()
fehler_do_while = df_vocabulary_do_while["Fehler"].dropna()

tasks_if = []
tasks_for = []
tasks_while = []
tasks_do_while = []


for fehler in fehler_if:
    fehler = str(fehler)
    tmp = [int(n) for n in fehler.split()]
    for x in tmp:
        tasks_if.append(x)

for fehler in fehler_for:
    fehler = str(fehler)
    tmp = [int(n) for n in fehler.split()]
    for x in tmp:
        tasks_for.append(x)
        
for fehler in fehler_while:
    fehler = str(fehler)
    tmp = [int(n) for n in fehler.split()]
    for x in tmp:
        tasks_while.append(x)
        
for fehler in fehler_do_while:
    fehler = str(fehler)
    tmp = [int(n) for n in fehler.split()]
    for x in tmp:
        tasks_do_while.append(x)

counter_fehler_if=collections.Counter(tasks_if)
counter_fehler_for=collections.Counter(tasks_for)
counter_fehler_while=collections.Counter(tasks_while)
counter_fehler_do_while=collections.Counter(tasks_do_while)

counter_fehler_if.most_common()
counter_fehler_for.most_common()
counter_fehler_while.most_common()
counter_fehler_do_while.most_common()

tmp = dict(sorted(counter_fehler_do_while.items(), key=lambda item: item[1], reverse=True))
for k,v in tmp.items():
    print("Nummer:", k, " |Anzahl:", v)
#tmp = collections.OrderedDict(sorted(counter_fehler_do_while.items()))
#for k, v in tmp.items():
#    print(k, v)
#print("For", counter_fehler_for)
#print("While", counter_fehler_while)
#print("Do-while", counter_fehler_do_while)
```
